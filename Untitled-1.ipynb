## Split the dataset to train and test.

```python
# Set the source directory
src_dir =  "/home/pouria/Data for test"

# Set the destination directories for train and test sets
train_dir = '/home/pouria/train_Fulhaüs'
test_dir = '/home/pouria/test_Fulhaüs'

# Set the split ratio for train and test sets
split_ratio = 0.8

# Create the train and test directories if they don't exist
os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

# Loop through each class directory in the source directory
for class_dir in os.listdir(src_dir):
    # Create the class directories in the train and test directories
    os.makedirs(os.path.join(train_dir, class_dir), exist_ok=True)
    os.makedirs(os.path.join(test_dir, class_dir), exist_ok=True)

    # Get the list of image files for this class
    image_files = os.listdir(os.path.join(src_dir, class_dir))

    # Shuffle the list of image files
    random.shuffle(image_files)

    # Split the image files into train and test sets
    split_index = int(len(image_files) * split_ratio)
    train_files = image_files[:split_index]
    test_files = image_files[split_index:]

    # Copy the train files to the train directories
    for train_file in train_files:
        src_file = os.path.join(src_dir, class_dir, train_file)
        dst_file = os.path.join(train_dir, class_dir, train_file)
        shutil.copyfile(src_file, dst_file)

    # Copy the test files to the test directories
    for test_file in test_files:
        src_file = os.path.join(src_dir, class_dir, test_file)
        dst_file = os.path.join(test_dir, class_dir, test_file)
        shutil.copyfile(src_file, dst_file)
```

## Data Visualization

```python
def plot_random_images(image_dir):
    """
    Plots six random images from each of the three classes (Bed, Chair, Sofa)
    in the given image directory.

    Args:
        image_dir (str): Path to the directory containing the image folders.
    """
    class_names = ['Bed', 'Chair', 'Sofa']
    fig, axs = plt.subplots(3, 2, figsize=(8, 10))

    for i, class_name in enumerate(class_names):
        class_path = os.path.join(image_dir, class_name)
        images = os.listdir(class_path)
        random.shuffle(images)
        for j in range(2):
            img_path = os.path.join(class_path, images[j])
            img = mpimg.imread(img_path)
            axs[i, j].imshow(img)
            axs[i, j].axis('off')
            axs[i, j].set_title(class_name)

    plt.tight_layout()
    plt.show()
```

## Image data generation

```python
train_IDG = ImageDataGenerator(rescale=1/255.,
                               rotation_range=20, # rotate the image slightly between 0 and 20 degrees
                               shear_range=0.2, # shear the image
                               zoom_range=0.2, # zoom into the image
                               width_shift_range=0.2, # shift the image width ways
                               height_shift_range=0.2, # shift the image height ways
                               horizontal_flip=True)
                                             
# We just rescale the test dataset
test_IDG = ImageDataGenerator(rescale=1/255.)
```

```python
train_data = train_IDG.flow_from_directory(train_dir, target_size=(224, 224),
                                               batch_size=32,class_mode='categorical')
                                               

test_data = test_IDG.flow_from_directory(test_dir,target_size=(224, 224),
                                    batch_size=32,class_mode='categorical')  
```

```python
# Create a function to implement a ModelCheckpoint callback with a specific filename 
def create_model_checkpoint(model_name, save_path="model_experiments"):
  return tf.keras.callbacks.ModelCheckpoint(filepath=os.path.join(save_path, model_name), # create filepath to save model
                                            verbose=0, # only output a limited amount of text
                                            save_best_only=True)
```

# Conv2D Model

```python
tf.random.set_seed(42) # random seed for Reproducibility
# Conv2D model 
model_1 = tf.keras.Sequential([
        tf.keras.layers.Conv2D(32, (2,2), activation='relu', input_shape=(224, 224,3)),
        tf.keras.layers.MaxPool2D(2,2),
        tf.keras.layers.Conv2D(64,  (2,2), activation='relu'),
        tf.keras.layers.MaxPool2D(2,2),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dropout(0.25),
        tf.keras.layers.Dense(512, activation='relu'),
        tf.keras.layers.Dense(3, activation='softmax') # softmax activation as the labels are multiclass
], name = "model1")

# early stopping callback to stop the model for avoiding overfitting.
early_stopping = tf.keras.callbacks.EarlyStopping(patience = 10) 

model_1.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
              loss='categorical_crossentropy', 
              metrics=['accuracy'])


history_1 = model_1.fit(train_data, epochs=80, verbose= 0,
                        validation_data = test_data,
                        callbacks=[create_model_checkpoint(model_name=model_1.name),early_stopping])
```

```python
model_1 = tf.keras.models.load_model("model_experiments/model1")
model_1_loss, model_1_accuracy = model_1.evaluate(test_data)
print(model_1_loss)
print(model_1_accuracy)
```

# Transfer Learning: MobileNetV3

```python
import tensorflow_hub as hub
IMAGE_SHAPE = (224, 224)
def create_model(model_url, num_classes=3):
  
  feature_extractor_layer = hub.KerasLayer(model_url,
                                           trainable=False, # freeze the underlying patterns
                                           name='feature_extraction_layer',
                                           input_shape=IMAGE_SHAPE+(3,)) # define the input image shape
  
  # Create our own model
  model = tf.keras.Sequential([
    feature_extractor_layer, # use the feature extraction layer as the base
    tf.keras.layers.Dense(num_classes, activation='softmax', name='output_layer') # create our own output layer      
  ])

  return model
```

```python
mobilenet_v3 = "https://tfhub.dev/google/imagenet/mobilenet_v3_large_100_224/classification/5"

mobilenet_v3_model = create_model(mobilenet_v3, num_classes=3)

# Compile
mobilenet_v3_model.compile(loss='categorical_crossentropy',
                     optimizer=tf.keras.optimizers.Adam(),
                     metrics=['accuracy'])
```

```python
mobilenet_v3_model.fit(train_data,
                        epochs=10,
                        steps_per_epoch=len(train_data),
                        validation_data=test_data,
                        validation_steps=len(test_data))
```

# FastAPI

For the API, We can use FastAPI to create a RESTful API that will allow us to receive image files and return predictions.


```python
import uvicorn
from fastapi import FastAPI, File, UploadFile
from PIL import Image
import io
import numpy as np
import tensorflow as tf

# Load the trained model
model = tf.keras.models.load_model('/home/pouria/best_model.h5')
```

```python
# Define the FastAPI app
app = FastAPI()

# Create a prediction endpoint
@app.post("/predict")
async def predict(file: UploadFile = File(...)):
    # Read the image file
    contents = await file.read()
    image = Image.open(io.BytesIO(contents))

    # Preprocess the image
    image = image.resize((224, 224))
    image = np.array(image) / 255.0
    image = image.reshape(1, 224, 224, 3)

    # Make prediction
    prediction = model.predict(image)
    class_idx = np.argmax(prediction[0])
    class_label = ['Bed', 'Chair', 'Sofa'][class_idx]

    # Return prediction result
    return {"class": class_label, "confidence": float(prediction[0][class_idx])}

if __name__ == "__main__":
    uvicorn.run(app, host="0.0.0.0", port=8000)

uvicorn app:app --reload
```

# Flask API

Create a Flask app to serve as the API endpoint

Load the trained model into the Flask app

Define a route to accept image data

Preprocess the image data to match the input format expected by your model

Use the model to predict the class of the image

Return the predicted class and its probability as a JSON response

```python
from flask import Flask, request, jsonify
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.applications.imagenet_utils import decode_predictions
import numpy as np
import cv2
import tensorflow_hub as hub

app = Flask(__name__)
model = load_model('/home/pouria/best_model.h5', custom_objects={'KerasLayer':hub.KerasLayer})

@app.route('/predict', methods=['POST'])
def predict():
    # Preprocess the image
    image = request.files.get('image').read()
    image = np.frombuffer(image, np.uint8)
    image = cv2.imdecode(image, cv2.IMREAD_COLOR)
    image = cv2.resize(image, (224, 224))
    image = image.astype('float') / 255.0
    image = img_to_array(image)
    image = np.expand_dims(image, axis=0)

    # Make predictions
    preds = model.predict(image)
    results = decode_predictions(preds, top=3)[0]
    predictions = []
    for result in results:
        label = result[1]
        probability = float(result[2])
        predictions.append({'label': label, 'probability': probability})
    return jsonify(predictions)

if __name__ == '__main__':
    app.run(debug=True, host='0.0.0.0', port=5000)
```

# Implement CI/CD pipeline on Github Actions
Create a .github/workflows directory in the repository

Create a main.yml file in the .github/workflows directory with the following contents:

```python
name: CI/CD Pipeline

on:
  push:
    branches:
      - main

jobs:
  build-and-test:
    runs-on: ubuntu-latest
    steps:
      - name: Checkout code
        uses: actions/checkout@v2

      - name: Build Docker image
        uses: docker/build-push-action@v2
        with:
          context: .
          push: false
          tags: myapp:latest

      - name: Test Docker image
        run: docker run -p 5000:5000 myapp:latest python test.py

  deploy:
    needs: build-and-test
    runs-on: ubuntu-latest
    steps:
      - name: Checkout code
        uses: actions/checkout@v2

      - name: Build and push Docker image
        uses: docker/build-push-action@v2
        with:
          context: .
          push
```
